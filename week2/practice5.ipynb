```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('img/sst-foyer.jpg')

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)


image2 = cv2.medianBlur(image, 15)

cv2.imwrite('img/median_blur.jpg', image2)

plt.subplot(1, 2, 2)
plt.title("Median Blur")
plt.imshow(image2)
plt.show()

```
