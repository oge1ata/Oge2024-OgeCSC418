```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('img/sst.jpg')

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)


image2 = cv2.resize(image, None, fx=2, fy=2)

cv2.imwrite('img/scaled.jpg', image2)

plt.subplot(1, 2, 2)
plt.title("Scaled")
plt.imshow(image2)
plt.show()

```
