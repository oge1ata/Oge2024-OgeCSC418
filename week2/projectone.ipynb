```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os

def addition(image1):
    image2 = cv2.imread('img/sst-foyer.jpg')

    image1 = cv2.resize(image1, (500, 400))
    image2 = cv2.resize(image2, (500, 400))

    addImage = cv2.addWeighted(image1, 0.5, image2, 0.6, 0)

    cv2.imshow('Weighted Image', addImage)

    if cv2.waitKey(0) & 0xff == 27:
        cv2.destroyAllWindows()

def contrast(image):
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)

    brightness = 5
    contrast = 1.5
    image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape, image.dtype), 0, brightness)

    cv2.imwrite('img/contrast_image.jpg', image2)

    plt.subplot(1, 2, 2)
    plt.title("Brightness and contrast")
    plt.imshow(image2)
    plt.show()

def sharpen(image):
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)

    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])

    image2 = cv2.filter2D(image, -1, kernel)

    cv2.imwrite('img/sharpened_image.jpg', image2)

    plt.subplot(1, 2, 2)
    plt.title("Sharpened Image")
    plt.imshow(image2)
    plt.show()

def medianBlur(image):
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)

    image2 = cv2.medianBlur(image, 15)

    cv2.imwrite('img/median_blur.jpg', image2)

    plt.subplot(1, 2, 2)
    plt.title("Median Blur")
    plt.imshow(image2)
    plt.show()

def resize(image):
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)


    image2 = cv2.resize(image, None, fx=2, fy=2)

    cv2.imwrite('img/scaled.jpg', image2)

    plt.subplot(1, 2, 2)
    plt.title("Scaled")
    plt.imshow(image2)
    plt.show()

def inverse(image):
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)


    image2 = 255 - image

    cv2.imwrite('img/inversed.jpg', image2)

    plt.subplot(1, 2, 2)
    plt.title("Inversed")
    plt.imshow(image2)
    plt.show()
```
