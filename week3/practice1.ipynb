```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

image = cv.imread('img/sst.jpg', 0)
rows, cols = image.shape

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)

M = np.float32([[1, 0, 100],[0, 1, 50]])
trans1 = cv.warpAffine(image, M, (cols, rows))


plt.subplot(1, 2, 2)
plt.title("Translated Image")
plt.imshow(trans1)
# plt.show()

cv.waitKey(0)
cv.destroyAllWindows()
```
