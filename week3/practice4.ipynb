```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

image = cv.imread('img/sst.jpg', 0)

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)


cropped = image[450:700, 200:500]


plt.subplot(1, 2, 2)
plt.title("Cropped Image")
plt.imshow(cropped)
plt.show()


cv.imwrite('img/cropped.jpg', cropped)
cv.waitKey(0)
cv.destroyAllWindows()
```
